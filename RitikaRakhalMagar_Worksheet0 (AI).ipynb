```python
#Task-1
def convert_units(conversion_type, value):
    """Converts units based on the selected conversion type."""

    if conversion_type == "1":  # meters to feet
        return value * 3.28084

    elif conversion_type == "2":  # feet to meters
        return value / 3.28084

    elif conversion_type == "3":  # kg to pounds
        return value * 2.20462

    elif conversion_type == "4":  # pounds to kg
        return value / 2.20462

    elif conversion_type == "5":  # liters to gallons
        return value * 0.264172

    elif conversion_type == "6":  # gallons to liters
        return value / 0.264172

    else:
        return None

def unit_converter():
    """Main function that asks the user for conversion type and value."""

    print("Unit Converter")
    print("Choose conversion type:")
    print("1. meters to feet")
    print("2. feet to meters")
    print("3. kilograms to pounds")
    print("4. pounds to kilograms")
    print("5. liters to gallons")
    print("6. gallons to liters")

    conversion_type = input("Enter option number: ")

    try:
        value = float(input("Enter value: "))

        result = convert_units(conversion_type, value)

        if result is None:
            print("Invalid conversion option")
        else:
            print("Converted value:", result)

    except ValueError:
        print("Invalid input. Please enter a number.")

unit_converter()
```

```python
#Task-2
def find_sum(numbers):
    """Returns the sum of numbers in the list."""
    return sum(numbers)

def find_average(numbers):
    """Returns the average of numbers in the list."""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns the maximum number in the list."""
    return max(numbers)

def find_min(numbers):
    """Returns the minimum number in the list."""
    return min(numbers)

def main():
    """Main function that asks the user to choose an operation and enter numbers."""

    print("Math Operations Program")
    print("1. Find Sum")
    print("2. Find Average")
    print("3. Find Maximum")
    print("4. Find Minimum")

    choice = input("Enter your choice: ")

    try:
        numbers_input = input("Enter numbers separated by spaces: ")

        numbers = [float(num) for num in numbers_input.split()]

        if len(numbers) == 0:
            print("Error: List cannot be empty")
            return

        if choice == "1":
            print("Sum:", find_sum(numbers))

        elif choice == "2":
            print("Average:", find_average(numbers))

        elif choice == "3":
            print("Maximum:", find_max(numbers))

        elif choice == "4":
            print("Minimum:", find_min(numbers))

        else:
            print("Invalid option")

    except ValueError:
        print("Error: Please enter valid numbers")

main()
```

Exercise on List Manipulation

```python
#1. Extract Every Other Element:
def extract_every_other(lst):
    """Return every other element from the list, starting with the first."""
    return lst[::2]

numbers = [1, 2, 3, 4, 5, 6]

result = extract_every_other(numbers)

print("Original list:", numbers)
print("Every other element:", result)
```

```python
#2. Slice a Sublist:
def get_sublist(lst, start, end):
    """
    This function returns a part of the list from start index to end index.
    """
    return lst[start:end+1]

numbers = [1, 2, 3, 4, 5, 6]
print(get_sublist(numbers, 2, 4))
```

```python
#3. Reverse a List Using Slicing:
def reverse_list(lst):
    """
    This function reverses the list using slicing.
    """
    return lst[::-1]

numbers = [1, 2, 3, 4, 5]
print(reverse_list(numbers))
```

```python
#4. Remove the First and Last Elements:
def remove_first_last(lst):
    """
    This function removes the first and last elements of a list.
    """
    return lst[1:-1]

numbers = [1, 2, 3, 4, 5]
print(remove_first_last(numbers))
```

```python
#5. Get the First n Elements:
def get_first_n(lst, n):
    """
    This function returns the first n elements of a list.
    """
    return lst[:n]

numbers = [1, 2, 3, 4, 5]
print(get_first_n(numbers, 3))
```

```python
#6. Extract Elements from the End:
def get_last_n(lst, n):
    """
    This function returns the last n elements of a list.
    """
    return lst[-n:]

numbers = [1, 2, 3, 4, 5]
print(get_last_n(numbers, 2))
```

```python
#7. Extract Elements in Reverse Order:
def reverse_skip(lst):
    """
    This function returns every second element from the list
    starting from the second last element going backwards.
    """
    return lst[-2::-2]

numbers = [1, 2, 3, 4, 5, 6]
print(reverse_skip(numbers))
```

Exercise on Nested List

```python
# 1. Flatten a Nested List
def flatten(lst):
    """Flatten a nested list into a single list."""
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(item)
        else:
            flat_list.append(item)
    return flat_list

nested_list1 = [[10, 20], [30, 40], [50]]
print("Flattened:", flatten(nested_list1))
```

```python
# 2. Accessing Nested List Elements
def access_nested_element(lst, indices):
    """Access an element in a nested list using a list of indices."""
    elem = lst
    for idx in indices:
        elem = elem[idx]
    return elem

nested_list2 = [[11, 22, 33], [44, 55, 66], [77, 88, 99]]
print("Accessed element:", access_nested_element(nested_list2, [1, 2]))
```

```python
# 3. Sum of All Elements in a Nested List
def sum_nested(lst):
    """Return the sum of all elements in a nested list, regardless of depth."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

nested_list3 = [[1, 3], [5, [7, 9]], 2]
print("Sum:", sum_nested(nested_list3))
```

```python
# 4. Remove Specific Element from a Nested List
def remove_element(lst, elem):
    """Remove all occurrences of an element from a nested list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

nested_list4 = [[3, 4], [4, 5], [6, 7]]
print("After removal:", remove_element(nested_list4, 4))
```

```python
# 5. Find the Maximum Element in a Nested List
def find_max(lst):
    """Find the maximum element in a nested list, regardless of depth."""
    max_val = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_val = max(max_val, find_max(item))
        else:
            max_val = max(max_val, item)
    return max_val

nested_list5 = [[2, 8], [10, [5, 12]], 7]
print("Maximum element:", find_max(nested_list5))
```

```python
# 6. Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    """Count the occurrences of an element in a nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

nested_list6 = [[1, 5], [5, 7], [8, 5]]
print("Occurrences of 5:", count_occurrences(nested_list6, 5))
```

```python
# 7. Flatten a List of Lists of Lists
def deep_flatten(lst):
    """Flatten a deeply nested list into a single list, regardless of depth."""
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list

nested_list7 = [[[1, 2], [3, 4]], [[5, 6], [7, 9]]]
print("Deep flattened:", deep_flatten(nested_list7))
```

```python
# 8. Nested List Average
def average_nested(lst):
    """Calculate the average of all elements in a nested list."""
    total_sum = sum_nested(lst)
    def count_elements(l):
        count = 0
        for item in l:
            if isinstance(item, list):
                count += count_elements(item)
            else:
                count += 1
        return count
    total_count = count_elements(lst)
    return total_sum / total_count if total_count != 0 else 0

nested_list8 = [[2, 4], [6, 8], [10, 12]]
print("Average:", average_nested(nested_list8))
```

Problem - 4: Matrix Operations:

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)
```

```python
# Compute inverse of A
A_inv = np.linalg.inv(A)

# Multiply A with its inverse
I_check = np.dot(A, A_inv)

print("A * A^-1:\n", I_check)

# Check if it's close to identity
print("Is A * A^-1 close to identity?", np.allclose(I_check, np.eye(2)))
```

```python
AB = np.dot(A, B)
BA = np.dot(B, A)

print("AB:\n", AB)
print("BA:\n", BA)

# Check if AB equals BA
print("Is AB equal to BA?", np.array_equal(AB, BA))
```

```python
AB_T = np.dot(A, B).T
BT_AT = np.dot(B.T, A.T)

print("(AB)^T:\n", AB_T)
print("B^T * A^T:\n", BT_AT)

# Check if they are equal
print("Is (AB)^T equal to B^T * A^T?", np.array_equal(AB_T, BT_AT))
```

```python
# Represent system in matrix form AX = B
A_sys = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])
B_sys = np.array([-1, -3, 9])

# Solve using np.linalg.solve (direct method)
X = np.linalg.solve(A_sys, B_sys)
print("Solution using linalg.solve:\n", X)
```

```python
# Compute inverse of A_sys
A_sys_inv = np.linalg.inv(A_sys)

# Solve X = A^-1 * B
X_inv = np.dot(A_sys_inv, B_sys)
print("Solution using inverse method:\n", X_inv)
```

Experiment: How Fast is Numpy?

```python
import numpy as np
import time
```

Step 1: Element-wise Addition

```python
def add_lists(list1, list2):
    start = time.time()
    result = [a + b for a, b in zip(list1, list2)]
    end = time.time()
    print("Python list addition time:", end - start, "seconds")
    return result

def add_arrays(arr1, arr2):
    start = time.time()
    result = arr1 + arr2
    end = time.time()
    print("NumPy array addition time:", end - start, "seconds")
    return result
```

Step 2: Element-wise Multiplication

```python
def multiply_lists(list1, list2):
    start = time.time()
    result = [a * b for a, b in zip(list1, list2)]
    end = time.time()
    print("Python list multiplication time:", end - start, "seconds")
    return result

def multiply_arrays(arr1, arr2):
    start = time.time()
    result = arr1 * arr2
    end = time.time()
    print("NumPy array multiplication time:", end - start, "seconds")
    return result
```

Step 3: Dot Product

```python
def dot_lists(list1, list2):
    start = time.time()
    result = sum(a * b for a, b in zip(list1, list2))
    end = time.time()
    print("Python list dot product time:", end - start, "seconds")
    return result

def dot_arrays(arr1, arr2):
    start = time.time()
    result = np.dot(arr1, arr2)
    end = time.time()
    print("NumPy array dot product time:", end - start, "seconds")
    return result
```

Step 4: Matrix Multiplication (1000x1000)

```python
def matmul_lists(mat1, mat2):
    size = len(mat1)
    start = time.time()
    result = [[sum(mat1[i][k] * mat2[k][j] for k in range(size))
               for j in range(size)] for i in range(size)]
    end = time.time()
    print("Python list matrix multiplication time:", end - start, "seconds")
    return result

def matmul_arrays(mat1, mat2):
    start = time.time()
    result = np.dot(mat1, mat2)
    end = time.time()
    print("NumPy array matrix multiplication time:", end - start, "seconds")
    return result
```

Run the functions

```python
size = 1_000_000
list1 = list(range(size))
list2 = list(range(size))
arr1 = np.arange(size)
arr2 = np.arange(size)

# Addition
add_lists(list1, list2)
add_arrays(arr1, arr2)

# Multiplication
multiply_lists(list1, list2)
multiply_arrays(arr1, arr2)

# Dot Product
dot_lists(list1, list2)
dot_arrays(arr1, arr2)

# Matrix Multiplication
matrix_size = 1000
mat1 = [[i+j for j in range(matrix_size)] for i in range(matrix_size)]
mat2 = [[i-j for j in range(matrix_size)] for i in range(matrix_size)]
np_mat1 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)
np_mat2 = np.arange(matrix_size**2).reshape(matrix_size, matrix_size)

matmul_lists(mat1, mat2)
matmul_arrays(np_mat1, np_mat2)
```
